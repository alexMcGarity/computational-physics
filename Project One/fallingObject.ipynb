```python
#calculates the height, velocity, and time of a falling object without air resistance
h = 50.0 #initial height in meters
g = 9.8 #acceleration due to gravity in m/s^2
v = 0. #initial velocity in m/s
t = 0. #initial time in seconds
dt = 0.1 #time step in seconds

while h >= 0. :
    h += v*dt
    v -= g*dt
    t += dt
    print("height = ", h, "meters")
    print("velocity = ", v, "m/s")
    print("time = ", t, "seconds")
```

```python
# calculates the height, velocity, and time of a falling object with air resistance
h = 50.0 #initial height in meters
g = 9.8 #acceleration due to gravity in m/s^2
v = 0. #initial velocity in m/s
t = 0. #initial time in seconds
dt = 0.1 #time step in seconds

while h >= 0. :
    a = -g + 0.01*v**2
    h += v*dt
    v += a*dt
    t += dt
print("height = ", h, "meters")
print("velocity = ", v, "m/s")
print("time = ", t, "seconds")
```
