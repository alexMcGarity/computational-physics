```python
import math
import random
import matplotlib.pyplot as plt

particleCount = 1000
particlesThrown = 0
particlesInCircle = 0
while particlesThrown < particleCount:
    particleX = random.uniform(-1, 1)
    particleY = random.uniform(-1, 1)
    if particleX**2 + particleY**2 <= 1:
        plt.scatter(particleX, particleY, color="0.0")
        particlesInCircle += 1
    else:
        plt.scatter(particleX, particleY, color="0.5")
    particlesThrown += 1

plt.title("Monte Carlo Simulation of a Circle")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()

print("Pi is approximately", 4 * particlesInCircle / particleCount)
```
