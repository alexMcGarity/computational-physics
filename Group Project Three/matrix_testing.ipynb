```python
import random
import matplotlib.pyplot as plt
import time
import matrix as m
import fastmatrix as fm

def matrix_inversion_test(n):
    "tests the runtime of the matrix inversion algorithm by plotting a log-log graph of runtime vs. matrix size up to n"
    x = []
    y = []
    for i in range(2, n):
        x.append(i)
        matrix = m.Matrix([[random.randint(-10, 10) for j in range(i)] for k in range(i)])
        start = time.time()
        matrix.inverse()
        end = time.time()
        y.append(end - start)
    # plot the data for matrix as a log graph
    plt.plot(x, y)
    plt.yscale("log")
    plt.xlabel("Matrix Size")
    plt.ylabel("Runtime (s)")
    plt.title("Matrix Inversion Runtime")
    plt.show()

    # plot the data for matrix as a log-log graph
    print("Log-Log Graph:")
    plt.plot(x, y)
    plt.yscale("log")
    plt.xscale("log")
    plt.xlabel("Matrix Size")
    plt.ylabel("Runtime (s)")
    plt.title("Matrix Inversion Runtime")
    plt.show()

matrix_inversion_test(9)
```

```python
import random
import matplotlib.pyplot as plt
import time
import matrix as m
import fastmatrix as fm

def fastmatrix_inversion_test(n):
    "tests the runtime of the matrix inversion algorithm by plotting a log-log graph of runtime vs. matrix size up to n"
    x = []
    y = []

    for i in range(2, n):
        x.append(i)
        matrix = fm.Matrix([[random.randint(-10, 10) for j in range(i)] for k in range(i)])
        start = time.time()
        matrix.inverse()
        end = time.time()
        y.append(end - start)
    # plot the data for matrix as a log graph
    print("Log Graph:")
    plt.plot(x, y)
    plt.yscale("log")
    plt.xlabel("Matrix Size")
    plt.ylabel("Runtime (s)")
    plt.title("Matrix Inversion Runtime")
    plt.show()
    # plot the data for matrix as a log-log graph
    print("Log-Log Graph:")
    plt.plot(x, y)
    plt.yscale("log")
    plt.xscale("log")
    plt.xlabel("Matrix Size")
    plt.ylabel("Runtime (s)")
    plt.title("Matrix Inversion Runtime")
    plt.show()

fastmatrix_inversion_test(100)
```
