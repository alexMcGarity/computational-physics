```python
import random


def matrix_inversion_test(n):
    "tests the runtime of the matrix inversion algorithm by plotting a log-log graph of runtime vs. matrix size up to n"
    import matplotlib.pyplot as plt
    import time
    x = []
    y = []
    for i in range(2, n):
        x.append(i)
        m = Matrix([[random.randint(-10, 10) for j in range(i)] for k in range(i)])
        start = time.time()
        m.inverse()
        end = time.time()
        y.append(end - start)
    # plot the data as a log-log graph
    plt.plot(x, y)
    plt.yscale("log")
    plt.xlabel("Matrix Size")
    plt.ylabel("Runtime (s)")
    plt.title("Matrix Inversion Runtime")
    plt.show()

matrix_inversion_test(11)
```
