Random Walk Algorithm

```python
'''Random Walk Algorithm:
1. Start at the origin
2. Take a random step of length 1 in a random diagonally forward direction
3. Repeat step 2 until you reach a distance of n from the origin
4. Repeat steps 1-3 for 1000 trials
5. Plot the results
'''
import random
import matplotlib.pyplot as plt

# Define the random walk function
def randomWalk(n):
    cords = [(0, 0)]
    for i in range(n):
        lastX = cords[i][0]
        cords.append((lastX + random.randint(-1, 1), i+1))
    return cords

# Define a function to run the random walk function multiple times
def runTrials(n, trials):
    allCords = []
    for i in range(trials):
        allCords.append(randomWalk(n))
    return allCords

# Define a function to plot the aggregate results of the random walk function on a histogram (x-axis is distance from origin, y-axis is number of trials)
def plotTrials(n, trials):
    allCords = runTrials(n, trials)
    x = []

    for i in range(len(allCords)):
        x.append(allCords[i][-1][0])

    plt.hist(x, bins=range(-25, 25))
    plt.title("Random Walk Results")
    plt.xlabel("Distance from Origin")
    plt.ylabel("Number of Trials")
    plt.show()
```

```python
'''execute the code for n = 100 and 100 trials'''
plotTrials(100, 100)
```

```python
'''execute the code for n = 1000 and 100 trials'''
plotTrials(1000, 100)
```

```python
'''execute the code for n = 100 and 1000 trials'''
plotTrials(100, 1000)
```
